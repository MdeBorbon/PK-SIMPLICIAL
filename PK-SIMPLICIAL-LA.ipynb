```python
def find_multiple_points(wiring):
    """
    Returns the points of intersection of the line arrangement from its wiring diagram.
    The output is two lists of tuples, first list is of double points and second list 
    The elements of the tuple are the lines that go through the point. 
    """

    # Number of lines of the arrangement
    n = len(wiring[0])
    
    # wiring diagram
    sequence = wiring[1]

    current_order = [i for i in range(1,n+1)]
    vertex_points = []

    for crossing in sequence:
        start_pos, end_pos = crossing
        
        # 1. Identify which lines are involved based on current order
        # (start_pos, end_pos) refers to the lines currently in those positions.
        involved_lines = []
        for i in range(start_pos - 1, end_pos):
            involved_lines.append(current_order[i])
        
        # Store this point
        vertex_points.append(tuple(sorted(involved_lines)))
        
        # 2. Update the order (permutation) for the next step
        # The lines involved in the crossing reverse their relative order
        involved_lines_reversed = involved_lines[::-1]
        current_order[start_pos-1:end_pos] = involved_lines_reversed
        
    dpoints = [p for p in vertex_points if len(p)==2]
    tpoints = [p for p in vertex_points if len(p) > 2]

    return dpoints, tpoints
```

```python
def pk_matrix(n, dpoints, tpoints):
    
    pk_matrix = matrix(n+1,n)

    for i in range(1,n+1):
        
        ti = [p for p in tpoints if i in p]
        di = [p for p in dpoints if i in p]
        
        pk_matrix[i-1,i-1] = len(ti) - 1
        
        for j in range(i+1,n+1):
            if (i,j) in di:
                pk_matrix[i-1,j-1] = 1
                pk_matrix[j-1,i-1] = 1
            else:
                pk_matrix[i-1,j-1] = 0
                pk_matrix[j-1,i-1] = 0
        for j in range(n):
            pk_matrix[n,j] = 1
    return pk_matrix
```

```python
semi_pk = dict()
for k in mpoints_by_Gkey.keys():
    n = k[0]
    dp, tp = mpoints_by_Gkey[k]
    A = pk_matrix(n, dp, tp)
    b = vector(n*[1]+[3])
    try:
        x = A.solve_right(b)
        ker_basis = A.right_kernel().basis()
        semi_pk[k] = [len(ker_basis), ker_basis, x]
    except ValueError:
        continue
```

```python
def affine_space_intersection_polyhedra(p, V, A, lower, upper):
    """
    p     : vector in R^n
    V     : list of direction vectors in R^n
    A     : m x n matrix
    lower : vector in (R ∪ {-∞})^m
    upper : vector in (R ∪ {+∞})^m
    """
    A = matrix(QQ, A)
    p = vector(QQ, p)
    V = [vector(QQ, v) for v in V]

    n = len(p)
    k = len(V)

    Mp = A * p
    Mv = [A * v for v in V]

    # Build inequalities in parameter space
    ieqs = []
    for i in range(A.nrows()):
        coeffs = [Mv[j][i] for j in range(k)]

        if lower[i] != -Infinity:
            ieqs.append([-lower[i] + Mp[i]] + coeffs)

        if upper[i] != Infinity:
            ieqs.append([upper[i] - Mp[i]] + [-c for c in coeffs])

    # Parameter polyhedron
    S = Polyhedron(ieqs=ieqs)

    # Affine map Phi
    def Phi(t):
        return p + sum(t[j] * V[j] for j in range(k))

    zero = vector(QQ, [0]*n)

    # Push forward V-representation
    verts = [Phi(v.vector()) for v in S.vertices()]
    rays  = [Phi(r.vector()) - p for r in S.rays()]
    lines = [Phi(l.vector()) - p for l in S.lines()]

    # Image polyhedron in R^n
    X = Polyhedron(vertices=verts, rays=rays, lines=lines)

    return S, X
```

```python
def basis_vectors(n):
    basis_vectors = []
    for i in range(1,n+1):
        v = vector([0]*(i-1)+[1]+[0]*(n-i))
        basis_vectors.append(v)
    return basis_vectors
```

```python
def indicator_vector(S, n):
    return vector([1 if i+1 in S else 0 for i in range(n)])
```

```python
pk_many_sol = dict()

for k in semi_pk_many_sol.keys():
    
    p = semi_pk_many_sol[k][2]
    V = semi_pk_many_sol[k][1]
    
    n = k[0]
    
    tpoints = mpoints_by_Gkey[k][1]
    m = len(tpoints)
    
    lower = vector(n*[0]+m*[0])
    upper = vector(n*[1]+m*[2])
    
    tp_vectors = []
    for point in tpoints:
        tp_vectors.append(indicator_vector(point,n))
    
    A = matrix(basis_vectors(n) + tp_vectors)
    
    pk_many_sol[k] = affine_space_intersection_polyhedra(p, V, A, lower, upper)[1]
```
